```python
import pandas as pd
data = pd.read_csv("documents/BYMA.BA.csv")
data.tail()
```

```python
#mg_hi = na(mg_hi[1]) ? ema(high, length) : mg_hi[1] + (high - mg_hi[1]) / (length * pow(high/mg_hi[1], 4))
```

```python
data['MG_HI']=data['High']
data['MG_LO']=data['Low']
data.tail()
```

```python
#for i in range(1, len(df)):
#    df.loc[i, 'C'] = df.loc[i-1, 'C'] * df.loc[i, 'A'] + df.loc[i, 'B']
for i in range(1, len(data)):
    data.loc[i,'MG_HI'] = data.loc[i-1,'MG_HI'] + (data.loc[i,'High']-data.loc[i-1,'MG_HI']) / (6*pow(data.loc[i,'High']/data.loc[i-1,'MG_HI'], 4))
    data.loc[i,'MG_LO'] = data.loc[i-1,'MG_LO'] + (data.loc[i,'Low']-data.loc[i-1,'MG_LO']) / (6*max(pow(data.loc[i,'Low']/data.loc[i-1,'MG_LO'], 4),0.3))
```

```python
data.tail()
```

```python
data['MG_HI']=data['MG_HI'].shift(4).ffill()
data['MG_LO']=data['MG_LO'].shift(4).ffill()
```

```python
data
```

```python
data = data.iloc[4:]
```

```python
data.reset_index(inplace=True,drop=True)
```

```python
data
```

```python
data['signal']=" "
for i in range (1,len(data)):
    if (data.loc[i,'Close']>data.loc[i,'MG_HI'] and data.loc[i-1,'Close']<data.loc[i-1,'MG_HI']):
        data.loc[i,'signal']="BUY"
    if (data.loc[i,'Close']<data.loc[i,'MG_LO'] and data.loc[i-1,'Close']>data.loc[i-1,'MG_LO']):
        data.loc[i,'signal']="SELL"
```

```python
data
```

```python
trading=pd.DataFrame()
trading=data[(data['signal']=="SELL") | (data['signal']=="BUY")]
trading
```

```python
trading.reset_index(drop=True,inplace=True)
trading
```

```python
trading['control']=' '
for i in range (1,len(trading)):
    if (trading.loc[i,'signal']==trading.loc[i-1,'signal']):
        trading.loc[i,'control']="delete"
trading
```

```python
trading.drop(trading[trading['control']=="delete"].index,axis=0,inplace=True)
trading
```

```python
trading.reset_index(drop=True,inplace=True)
trading.drop(columns='control',inplace=True)
trading
```

```python
trading['cum']=0
for i in range(0,len(trading)):
    if(i == 0 ):
         trading.loc[i,'cum']= trading.loc[i,'Close'] if(trading.loc[i,'signal']=="SELL") else trading.loc[i,'Close']*-1
    else:
        trading.loc[i,'cum']=trading.loc[i-1,'cum']-trading.loc[i,'Close'] if(trading.loc[i,'signal']=="BUY") else trading.loc[i-1,'cum']+trading.loc[i,'Close']
```
